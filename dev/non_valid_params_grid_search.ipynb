```python
from skforecast.recursive import ForecasterRecursive
from skforecast.model_selection import grid_search_forecaster, TimeSeriesFold
from sklearn.linear_model import ElasticNet
import numpy as np
import pandas as pd
import warnings

# Simulate time series
np.random.seed(123)
n = 200
x = np.arange(n)
y = 0.5 * x + 10 * np.sin(x) + 50 * np.random.normal(size=n)
y = pd.Series(y)
y.index = pd.date_range(start='2000-01-01', periods=n, freq='D')

param_grid = {
    'alpha': [0.01, 0.1, 1.0],
    'l1_ratio': [0.0, 0.5, 10]
}

cv = TimeSeriesFold(
         steps              = 12,
         initial_train_size = 100,
         refit              = False
     )

forecaster = ForecasterRecursive(
                 regressor = ElasticNet(),
                 lags      = 10
             )


results = grid_search_forecaster(
              forecaster    = forecaster,
              y             = y,
              param_grid    = param_grid,
              cv            = cv,
              metric        = 'mean_squared_error',
              return_best   = True,
              n_jobs        = 'auto',
              verbose       = False,
              show_progress = True
          )
results
```
