```python
import numpy as np
import pandas as pd
import skforecast
from sklearn.linear_model import LinearRegression
from skforecast.direct import ForecasterDirect, ForecasterDirectMultiVariate
from skforecast.preprocessing import RollingFeatures
from skforecast.utils import exog_to_direct, exog_to_direct_numpy
import time
import warnings
from skforecast.exceptions import DataTransformationWarning

warnings.simplefilter("ignore", category=DataTransformationWarning)
```

```python
series = pd.DataFrame(
    {
        f"series_{i}": pd.Series(
            np.random.randn(1000),
            index=pd.date_range(start="2000-01-01", periods=1000, freq="D"),
        )
        for i in range(10)
    }
)

y = pd.Series(
    np.random.randn(5000),
    index=pd.date_range(start="2000-01-01", periods=5000, freq="D"),
)

exog = pd.DataFrame(
    np.random.randn(5000, 50),
    index=pd.date_range(start="2000-01-01", periods=5000, freq="D"),
    columns=[f"exog_{i}" for i in range(50)],
)
```

```python
forecaster = ForecasterDirect(estimator=LinearRegression(), lags=48, steps=24)

start = time.perf_counter()

forecaster.fit(y=y, store_in_sample_residuals=True)

end = time.perf_counter()
print(f"Time taken: {end - start} seconds")
```

```python
window_features = RollingFeatures(window_sizes=48, stats=["mean", "std", "min", "max"])
forecaster = ForecasterDirect(
    estimator=LinearRegression(), lags=48, window_features=window_features, steps=24
)

start = time.perf_counter()

forecaster.fit(y=y, store_in_sample_residuals=True)

end = time.perf_counter()
print(f"Time taken: {end - start} seconds")
```

```python
forecaster = ForecasterDirectMultiVariate(
    estimator=LinearRegression(),
    lags=48,
    steps=10,
    level="series_1",
)

start = time.perf_counter()

forecaster.fit(series=series, store_in_sample_residuals=True)

end = time.perf_counter()
print(f"Time taken: {end - start} seconds")
```
